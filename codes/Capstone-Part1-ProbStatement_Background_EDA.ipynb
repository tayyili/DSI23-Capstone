## Problem Statement

With earlier care and attention given, mortality by heart failure can be prevented. Thus, the hospital's newly established Data Science Department has been tasked by the Department of Cardiology to find a way to identify patients with high risks of mortality by heart failure through the use of data science, so as to provide the necessary preventive care and attention early.

To achieve this, the project aims to build a classifier which uses patients' existing health conditions to accurately predict mortality by heart failure.

## Background

The heart has 4 chambers (right and left atriums on top, right and left ventricles below) in which blood flows through [(source)](https://medlineplus.gov/ency/imagepages/19612.htm). 

The right atrium receives blood from the veins and pumps it into the right ventricle. The right ventricle then pumps the blood into the lungs to fill the blood with oxygen.

The left atrium receives the oxygenated blood from the lungs and pumps it into the left ventricle (which is usually the strongest chamber). The left ventricle then pumps the oxygenated blood to the rest of the body [(source)](https://www.webmd.com/heart/picture-of-the-heart).

When the heart contracts, blood is pumped out from the ventricles and when the heart relaxes, the ventricles are refilled with blood from the atriums. There will always be some blood left in the ventricles as blood can never be fully pumped out of the ventricles no matter how forceful the heart contracts [(source)](https://www.mayoclinic.org/ejection-fraction/expert-answers/faq-20058286). 

Heart failure occurs when the heart muscle is unable to pump blood as well as it should and the patient does not get enough oxygen [(source)](https://www.heart.org/en/health-topics/heart-failure/what-is-heart-failure). Heart failure has no cure and can be fatal if not detected and managed early [(source)](https://www.heart.org/en/health-topics/heart-failure/treatment-options-for-heart-failure) [(source)](https://www.mayoclinic.org/diseases-conditions/heart-failure/symptoms-causes/syc-20373142).

There are many different types of heart failure, with the most common being [(source)](https://my.clevelandclinic.org/health/diseases/17069-heart-failure-understanding-heart-failure):

|Type of Heart Failure|Description|
|:--|:--|
|Heart Failure with Reduced Ejection Fraction (**HF-rEF**)|Left ventricle enlarges, becoming unable to contract normally and as a result, the heart fails to pump with enough force to push sufficient amount of oxygenated blood into circulation around the body|
|Heart Failure with Preserved Ejection Fraction (**HF-pEF**)|Heart pumps normally but the ventricles are unable to relax properly as the muscles are thicker and stiffer than usual. This results in ventricles not getting filled all the way, causing lesser blood to be pumped to the rest of the body|

A frequent cause of heart failure is cardiovascular diseases (CVDs) [(source)](https://world-heart-federation.org/cvd-roadmaps/whf-global-roadmaps/heart-failure/). CVDs have consistently been a leading cause of death globally [(source)](https://world-heart-federation.org/resource/cardiovascular-diseases-cvds-global-facts-figures/). In 2019, it took an estimate of 17.9 million lives, which accounts to about 32% of all deaths that year [(source)](https://www.who.int/news-room/fact-sheets/detail/cardiovascular-diseases-(cvds)). 

Fortunately, most CVDs can be prevented by addressing various behavioural risk factors in everyday life like unhealthy diet and obesity, tobacco use and lack of physical exercise [(source)](https://www.mayoclinic.org/diseases-conditions/heart-failure/symptoms-causes/syc-20373142). 

Patients with other underlying conditions such as high blood pressure are also at high risk of developing CVDs [(source)](https://www.cdc.gov/globalhealth/healthprotection/ncd/cardiovascular-diseases.html). 

Therefore, for patients with CVDs or are at high risk of CVDs, early detection and management of these risk factors and underlying conditions is key to preventing mortality by heart failure [(source)](https://www.who.int/news-room/fact-sheets/detail/cardiovascular-diseases-(cvds)). 

The dataset used in this project contains 299 patients with heart failure and was obtained from the research article ["Machine learning can predict survival of patients with heart failure from serum creatinine and ejection fraction alone"](https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5#Abs1) (Chicco & Jurman, 2020).

### Background of Features

#### Anaemia Background

Anaemia is a condition where the patient has lower than usual red blood cells level (hematocrit) or hemoglobin concentration. A reduction in hematocrit or hemoglobin concentration reduces the body's capacity in carrying the required amount of oxygen to the body's tissues, which can cause the patient to feel weak, fatigue, dizzy, etc [(source)](https://www.who.int/health-topics/anaemia#tab=tab_1).

There are many different forms of anaemia [(source)](https://www.radiologyinfo.org/en/info/anemia):

|Types of Anaemia|Description|
|:--|:--|
|Iron Deficiency Anaemia|Low iron levels in blood, which is required by the body to create hemoglobin|
|Vitamin Deficiency Anaemia|Low levels of vitamins C, B-12 or folate in blood|
|Aplastic Anaemia|Bone marrow unable to produce sufficient red blood cells|
|Hemolytic Anaemia|Red blood cells destroyed prematurely by body|
|Sickle Cell Anaemia|Inherited disorder where red blood cells have an abnormal crescent shape|
|Thalassemia|Inherited disorder where red blood cells are destroyed prematurely by an abnormal form of hemoglobin|

Anaemia is usually first diagnosed through a Complete Blood Cell (CBC) test to measure the amount of various components in the blood, which includes red blood cells and hemoglobin [(source)](https://www.mayoclinic.org/diseases-conditions/anemia/diagnosis-treatment/drc-20351366). The normal range for hematocrit and hemoglobin concentration varies from person to person depending on things like age and also varies slightly among medical practices, but the table below shows the general normal range [(source)](https://www.redcrossblood.org/donate-blood/dlp/hematocrit.html):

|Component|Normal Range for Men|Normal Range for Women|
|:--|:--|:--|
|Hematocrit (percentage of red blood cells in blood)|41% to 50%|36% to 48%|
|Hemoglobin Concentration (enables red blood cells to transport oxygen, in g/dL)|13.5g/dL to 17.5g/dL|12.0g/dL to 15.5g/dL|

Values lower than the normal range indicates that the patient is anaemic.

Complications of anaemia if left untreated includes [(source)](https://www.mayoclinic.org/diseases-conditions/anemia/symptoms-causes/syc-20351360):
- Heart problems (rapid or irregular heartbeat. Heart pumps more blood to make up for reduced oxygen supply to the rest of the body, potentially leading to enlarged heart or heart failure)
- Death (Inherited anaemias like Sickle Cell Anaemia can lead to life threatening complications. Losing too much blood too quickly results in severe anaemia, which is fatal)

*In the dataset used, patients are classified based on whether they have anaemia (denoted as 1) or not (denoted as 0).*

#### Creatine Phosphokinase Background

Creatine Phosphokinase (also refered to as CPK, Creatine Kinase or CK) is an enzyme in the body that is found in the heart, brain and skeletal muscles. CPK leaks into the blood when the muscle tissue is damaged [(source)](https://www.hopkinslupus.org/lupus-tests/clinical-tests/creatine-phosphokinase-cpk/).

Total CPK level is made up of 3 different forms of CPK added together. Each of these forms have slightly different substances and are found mostly in different areas of the body [(source)](https://www.ucsfbenioffchildrens.org/medical-tests/cpk-isoenzymes-test). The specific form of CPK that is found to be at a higher than normal level gives an indication of which area of the body is injured/facing abnormalities and therefore, what condition the patient is suffering from [(source)](https://www.portea.com/labs/diagnostic-tests/creatine-phosphokinase-cpk-ck-mb-bb-mm-83/).

|Specific Forms of CPK|Percentage of Total CPK [(source)](https://www.glowm.com/lab-text/item/25)|Where it is mostly found|Examples of Potential Conditions|
|:--:|:--:|:--:|:--|
|CPK-1 (also known as CPK-BB)|0%|Brain and Lungs|<ul><li>Brain Cancer</li><li>Brain Injury (including stroke or bleeding)</li><li>Seizure</li></ul>|
|CPK-2 (also known as CPK-MB)|0% to 3%|Heart|<ul><li>Heart Injury</li><li>Heart Inflammation</li><li>Heart Attack</li><li>Electrical Injury</li></ul>|
|CPK-3 (also known as CPK-MM)|97% to 100%|Skeletal Muscle|<ul><li>Muscle Inflammation</li><li>Muscle Crush Injury</li><li>Muscle Dystrophy</li><li>Muscle Trauma</li></ul>|

If a patient is found to have high total CPK level in the blood during a CPK test, it might be an indication of stress or injury to the heart or other muscles. The doctor may then conduct further tests to check for the specific form of CPK that is elevated in order to determine the source of damage [(source)](https://www.hopkinslupus.org/lupus-tests/clinical-tests/creatine-phosphokinase-cpk/).

As the normal values for total CPK and the specific forms of CPK varies depending on the patient in question and the labs conducting the tests, there is no absolute reference to compare against. However, a common reference range that seems to be used for CPK-3 (also known as CPK-MB) is that it **falls within 3% of the total CPK level** and is **approximately between 0 mcg/L to 5 mcg/L** [(source)](http://www.clinlabnavigator.com/creatine-kinase-mb-ck-mb.html) [(source)](http://thekahancenter.com/library/Item.aspx?HWID=hw224485&sec=hw224494).

*The value that is recorded and used in the dataset of this project is CPK-3 (also known as CPK-MB), recorded in mcg/L (1 ng/mL is equivalent to 1 mcg/L).*

#### Diabetes Background

Diabetes is a chronic disease which occurs when the body is unable to properly regulate the blood sugar. This inability to regulate blood sugar can be due to either the pancreas producing insufficient insulin or the body being unable to effectively use the insulin that it produces [(source)](https://www.who.int/news-room/fact-sheets/detail/diabetes).

Insulin can be thought of as a key that allows blood sugar into the body's cells for use as an energy source. The lack of insulin or the inability to effectively use insulin will cause the blood sugar to stay in the bloodstream in excess [(source)](https://www.cdc.gov/diabetes/basics/diabetes.html#:~:text=Diabetes%20is%20a%20chronic%20(long,your%20pancreas%20to%20release%20insulin.). If left uncontrolled over a long period of time, there will be serious damage to multiple areas of the body [(source)](https://www.healthline.com/health/diabetes#symptoms).

Complications arising from diabetes may include [(source)](https://www.mayoclinic.org/diseases-conditions/diabetes/symptoms-causes/syc-20371444):
- Cardiovascular disease (more likely to get heart disease or stroke)
- Nerve damage (numbness or pain at tips of toes or fingers, gradually spreading upwards and if left untreated, will lose all sense of feeling in the limb)
- Kidney damage (kidney failure or irreversible end-stage kidney disease requiring kidney transplant or dialysis)
- Eye damage (potentially lead to blindness)

There are multiple forms of diabetes, with the most common types being [(source)](https://www.niddk.nih.gov/health-information/diabetes/overview/what-is-diabetes) [(source)](https://my.clevelandclinic.org/health/diseases/7104-diabetes-mellitus-an-overview):

|Type of Diabetes|Description|
|:--|:--|
|Type 1 Diabetes|<ul><li>Immune system attacks cells in pancreas that produces insulin and destroys them</li><li>Need to take insulin everyday to survive as body no longer produces insulin</li></ul>|
|Type 2 Diabetes|<ul><li>Insulin produced by body is insufficient</li><li>Body's cells have become resistant to insulin and does not respond normally to the insulin</li></ul>|
|Gestational Diabetes|<ul><li>Develops in some women when pregnant</li><li>Hormones produced during pregnancy makes body cells more resistant to insulin</li><li>Usually goes away after pregnancy but at higher risk of developing Type 2 Diabetes in future</li></ul>|

*Patients are classified in the dataset based on whether they have diabetes (denoted as 1) or not (denoted as 0).*

#### Ejection Fraction Background

The percentage of blood pumped out of the heart's main pumping chamber (usually the left ventricle) at each heartbeat is called the Ejection Fraction (EF). It is an indication of how well the heart can pump out blood and is used to diagnose and track heart failure [(source)](https://www.heart.org/en/health-topics/heart-failure/diagnosing-heart-failure/ejection-fraction-heart-failure-measurement).

Different ranges of EF percentage gives different indications of how well the heart is working and what condition the heart is currently in  [(source)](https://my.clevelandclinic.org/health/articles/16950-ejection-fraction):

|EF Percentage|Pumping Ability of Heart|Level of Failure|
|:--|:--|:--|
|55% to 70%|Normal|Heart is either normal or patient may have [HF-pEF](#Background)|
|40% to 54%|Slightly below normal|Less blood available in heart so less blood pumped out. Lower-than-normal amount of oxygenated blood circulated to the rest of the body|
|35% to 39%|Moderately below normal|Mild [HF-rEF](#Background)|
|Below 35%|Severely below normal|Moderate-to-severe HF-rEF, with severe HF-rEF increasing risk of life-threatening heartbeats and having right and left ventricles pumping out-of-sync|

*Values recorded and used in the dataset of this project are in percentage.*

#### High Blood Pressure Background

The amount of blood that the heart pumps and the amount of resistance to the blood flow in the arteries (major blood vessels in the body) makes up blood pressure [(source)](https://www.healthline.com/health/high-blood-pressure-hypertension#overview). Blood flowing through the arteries puts pressure on the artery walls and this pressure is called the blood pressure [(source)](https://familydoctor.org/condition/high-blood-pressure/).

The higher the amount of blood pumped by the heart and the narrower the arteries, the higher the blood pressure [(source)](https://www.mayoclinic.org/diseases-conditions/high-blood-pressure/symptoms-causes/syc-20373410).

High Blood Pressure (HBP) is also known as hypertension, which indicates that the blood pressure measure is consistently higher than normal. Higher blood pressure results in higher risk of other health issues including heart diseases, heart attack and stroke [(source)](https://www.cdc.gov/bloodpressure/about.htm).

Blood pressure measurement consists of 2 values [(source)](https://www.who.int/news-room/fact-sheets/detail/hypertension):
- First value (the systolic number) represents pressure in blood vessels when heart contracts
- Second value (the diastolic number) represents pressure in blood vessels when heart is relaxed and resting between beats

The ideal blood pressure is usually between 90/60mmHg and 120/80mm/Hg [(source)](https://www.nhs.uk/conditions/high-blood-pressure-hypertension/). 

A patient is diagnosed with HBP if their blood pressure, measured on 2 different days, are both above 140/90mmHg [(source)](https://www.who.int/news-room/fact-sheets/detail/hypertension).

*Patients are classified in the dataset based on whether they have HBP (denoted as 1) or not (denoted as 0).*

#### Platelets Background

Platelets are small blood cells that stops bleeding at damaged areas by forming clots. When the platelets reaches the damaged area, they grow sticky tentacles that allows them to stick to one another while sending out chemical signals to attract more platelets, piling onto each other and eventually forming a clot [(source)](https://www.urmc.rochester.edu/encyclopedia/content.aspx?ContentTypeID=160&ContentID=36).

A routine blood test called the Complete Blood Count (CBC) will be able reveal the number of platelet counts in a patient. The normal platelet count ranges from 150,000 to 450,000 platelets per microliter. Having more than 450,000 is a condition known as thrombocytosis while having less than 150,000 is a condition known as thrombocytopenia [(source)](https://www.verywellhealth.com/platelet-count-plt-5192703).

Having too many platelets (thrombocytosis) can cause spontaneous blood clots in arms and legs, which if left untreated may lead to heart attack and stroke [(source)](https://www.hopkinsmedicine.org/health/conditions-and-diseases/what-are-platelets-and-why-are-they-important).

*Values used in the dataset of this project are recorded in kiloplatelets/mL.*

#### Serum Creatinine Background

Creatinine is a waste product produced by the muscles and is usually filtered out of the blood by the kidneys. They are released into urine and expelled from the body. Measuring the creatinine levels in the blood or urine gives a hint of how well the kidneys are functioning. [(source)](https://www.freseniuskidneycare.com/thrive-central/what-are-high-creatinine-levels).

The typical range for serum creatinine is [(source)](https://www.mayoclinic.org/tests-procedures/creatinine-test/about/pac-20384646):
- 0.74 to 1.35 mg/dL for adult men
- 0.59 to 1.04 mg/dL for adult women

Elevated levels of serum creatinine can be an indicator of poor kidney function which may be caused by kidney disease, which in turn may be caused by high blood pressure or diabetes [(source)](https://www.medicinenet.com/creatinine_blood_test/article.htm).

Heart failure might be another cause of kidney disease since the kidneys and heart works closely together. If the heart is unable to pump efficiently, the kidneys suffer from reduced supply of oxygenated blood. There is also congestion of blood due to buildup of pressure in the main vein connected to the kidneys. 

In an attempt to increase blood supply to the kidneys, the hormone system regulating blood pressure goes into overdrive. The heart, as a result, has an increase in workload to pump against higher pressure in the arteries which causes it to suffer eventually. 

Therefore, it is common to check for kidney function through serum creatinine levels if heart disease is suspected for a patient [(source)](https://www.bhf.org.uk/informationsupport/heart-matters-magazine/medical/kidney-heart-link).

*Values used in the dataset of this project are recorded in mg/dL.*

#### Serum Sodium Background

Sodium is a type of electrolyte (electrically charged minerals) and is essential to helping nerves and muscles function properly. Most of the sodium required by the body is gotten from our diet. Excess sodium is filtered and removed by the kidney through urine [(source)](https://medlineplus.gov/lab-tests/sodium-blood-test/).

Serum sodium, also known as sodium blood, is the value of sodium inside the blood. A sodium blood test is conducted to check for the amount of sodium in the blood [(source)](https://www.healthline.com/health/sodium-blood). 

The normal range for sodium blood levels is 135 to 145 mEq/L [(source)](https://www.ucsfhealth.org/medical-tests/sodium-blood-test). If the sodium blood level in the body drops below 135 mEq/L, a disorder called hyponatremia occurs, which can be fatal if too severe. Symptoms for hyponatremia may include [(source)](https://www.medicalnewstoday.com/articles/323831):
- sluggishness
- headache
- fatigue and low energy
- nausea
- vomitting (severe)
- muscle weakness, spasms, twitching (severe)
- seizures (severe)

Drop in serum sodium levels leading to hyponatremia can be caused by many different reasons, including heart, kidney and liver problems. 

Congestive heart failure and other such diseases affect the kidneys or liver, causing an accumuluation of fluid in the body. This dilutes the sodium in the body and lowers the overall serum sodium level, leading to hyponatremia [(source)](https://www.mayoclinic.org/diseases-conditions/hyponatremia/symptoms-causes/syc-20373711).

*Values used in the dataset of this project are recorded in mEq/L.*

### Heatmap

```python
#heatmap cell size
plt.figure(figsize = (18, 8))

#mask upper triangle of heatmap
corr = data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

#cmap adjustment so negative and positive values are reflected with the same color
cmap = sns.diverging_palette(1, 1, as_cmap=True)

#plotting heatmap to check which features are likely to have a bigger influence on our final predictions
heatmap = sns.heatmap(corr, annot=True, cmap=cmap, mask=mask);

heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=50);

plt.savefig('../plots/heatmap.png', bbox_inches="tight", dpi=400)

plt.show();
```

Based on the background research done, it was expected that most of the features in the dataset would have high correlation with the target variable (death_event) as the features either results directly or indirectly to heart failure.

However, from the heatmap analysis and focusing on the values on the bottom-most row, we see that most of the features actually have negligible correlation with the target variable. 

Only the following features seem to have some correlation with the target variable (taking features with correlation value above absolute 0.25):

- age
- ejection_fraction
- serum_creatinine
- time

Therefore, we will only be using the features mentioned above in our modelling to avoid introducing too much noise and cause potential overfitting.
